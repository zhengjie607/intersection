```python
from CLI_Reader import CLI
from intersection import intersection
import numpy as np
import matplotlib.pyplot as plt
import math
def Rotate_matrix(angle,cenx,ceny):
    C=math.cos(angle*math.pi/180)
    S=math.sin(angle*math.pi/180)
    M11=C
    M12=-S
    M13=(1-C)*cenx+S*ceny
    M21=S
    M22=C
    M23=(1-C)*ceny-S*cenx
    return np.array([[M11,M12,M13],[M21,M22,M23],[0,0,1]])
def Scale_matrix(sx,sy,cenx,ceny):
    M11=sx
    M13=(1-sx)*cenx
    M22=sy
    M23=(1-sy)*ceny
    return np.array([[M11,0,M13],[0,M22,M23],[0,0,1]])
def scalebound(scale,Xmin,Xmax,Ymin,Ymax):
    cenx=(Xmin+Xmax)/2
    ceny=(Ymin+Ymax)/2
    return (Xmin-cenx)*scale+cenx,(Xmax-cenx)*scale+cenx,(Ymin-ceny)*scale+ceny,(Ymax-ceny)*scale+ceny
def getPoint(cli,Angle,Distance):
    bound_Xmin,bound_Xmax,bound_Ymin,bound_Ymax=scalebound(1.5,cli.Xmin,cli.Xmax,cli.Ymin,cli.Ymax)
    '''bound_Xmin=0
    bound_Xmax=10
    bound_Ymin=0
    bound_Ymax=10'''
    cenx=(bound_Xmin+bound_Xmax)/2
    ceny=(bound_Ymin+bound_Ymax)/2
    print(cenx,ceny)
    step=int((bound_Ymax-bound_Ymin)/Distance)+1
    w=[]
    for i in range(step):
        w.append([bound_Xmin,bound_Ymin+Distance*i])
        w.append([bound_Xmax,bound_Ymin+Distance*i])
    one=np.ones(len(w))
    w=np.array(w)
    c=np.insert(w,2,values=one.T,axis=1)
    '''x1=c[:,0]
    y1=c[:,1]
    plt.plot(x1,y1,c='r')
    print(c)'''
    M=Rotate_matrix(Angle,cenx,ceny)
    result=np.dot(c,M.T)
    '''x2=result[:,0]
    y2=result[:,1]
    plt.plot(x2,y2,c="g")
    print(result)
    plt.show()'''
    for i in range(int(len(result)/2)):
        x2=result[2*i:2*i+2,0]
        y2=result[2*i:2*i+2,1]
        #plt.plot(x2,y2,c="g")
        e=cli.UsePoint[0]
        p=[]
        for w in e:
            w=np.array(w)
            x1=w[:,0]
            y1=w[:,1]
            pp=intersection(x1,y1,x2,y2) 
            if len(pp)>=2:
                
                p.append(pp)
                print('point:',p)
                print('len of point:',len(p))
            '''if len(p)>=2:
                print(p)
                print(p[:,0])
                plt.plot(p[:,0],p[:,1],c='g')'''
    plt.show()
if __name__=='__main__':
    path=r"F:\qp\三个孔.cli"
    cli=CLI(path)
    getPoint(cli,45,10)
```
